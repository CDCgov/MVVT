**Purpose:** This notebook identifies BCR sequences containing reported influenza-associated sequence motifs. The search is performed using heavy- and light-chain V/J gene calls, CDR3 and/or CDR2 amino acid sequences, and translated nucleotide sequences where needed.

**Input:** A tab-delimited file containing AIRR-compatible BCR annotations. Heavy-chain columns should be prefixed with IGH_, and light-chain columns should be prefixed with IGL_.

**Output:** A tab-delimited file containing antibody sequences with at least one predicted influenza-specific motif assignment.

```python
import Bio
from Bio import SeqIO,Seq
from pathlib import Path
from Bio.SeqRecord import SeqRecord
import pandas as pd
import numpy as np
import re
```

```python
folder = '' 
tsvF = 'Input_file_for_motif_search.tsv'         # Input file as tab-separated CSV in AIRR format
```

```python
## This script works with AIRR-formatted files. Files in other formats may require column renaming. 
# Heavy- and light-chain column names should be distinguished by adding IGH_ and IGL_ prefixes, respectively.
df = pd.read_csv(Path(folder, tsvF),sep='\t',low_memory=False)
```

```python
def check_columns(df): 
    ## Required-column check: ##
    required_cols = ['IGH_v_call', 'IGH_d_call', 'IGH_j_call', 'IGH_sequence', 'IGH_junction_aa',
                     'IGL_v_call', 'IGL_j_call', 'IGL_sequence', 'IGL_junction_aa' ]
    igl_cols = { 'IGL_v_call','IGL_j_call','IGL_sequence','IGL_junction_aa' }

    # Initial check
    missing_cols = [col for col in required_cols if col not in df.columns]
    if missing_cols: 
        print("Missing columns:", missing_cols, "\n") 
        # Missing only IGL columns 
        if set(missing_cols).issubset(igl_cols):
            print('''
It appears that your input file is missing some IGL columns.

Since some reported influenza-specific BCR signatures depend on both heavy- and light-chain sequence features, 
datasets containing only unpaired heavy chains should include blank values for the required light-chain fields.

We will create empty IGL columns to allow the system to run. \n''')

            df = create_light_chains(df)

            # Re-check after fixing
            missing_cols = [col for col in required_cols if col not in df.columns ]

            if missing_cols:
                raise ValueError("Required columns are still missing after repair: %s" % ', '.join(missing_cols))
        else:
            raise ValueError("Required columns are missing: %s" % ', '.join(missing_cols))
    print('All required columns are present')
    return df


def create_light_chains(df):
    # For datasets containing unpaired heavy chains,
    # assign blank values for required light-chain fields.

    print('Assigned blank values for the required light-chain fields')
    for col in ['IGL_v_call', 'IGL_j_call', 'IGL_sequence', 'IGL_junction_aa']:
        if col not in df.columns:
            df[col] = ""
    return df

# Translate nucleotide sequences in all reading frames and return the longest translated ORF. 
def translate_by_3s(seq):
    orfs = []
    for s in (seq, seq.reverse_complement()):
        s = str(s)
        for  frame in range(3):
            s_frame = s[frame:]
            while len(s_frame)%3 != 0:
                s_frame += 'N'
            orfs.extend((Seq.Seq(s_frame).translate(to_stop=False)).split('*'))
            
    return max(orfs, key=len)
```

```python
df = check_columns(df)
```

```python
# In the input dataset some records may be missing CDR3 sequences. Replace missing CDR3 values with blank strings.
df['IGH_junction_aa'] = df['IGH_junction_aa'].fillna('')
df['IGL_junction_aa'] = df['IGL_junction_aa'].fillna('')
```

### Functions to check for each motif

```python
def check_IGHV1_69_HAstem_binding(cdr3,vhseq):
    # Check for conserved residues in IGH1-69 HCDR3 sequences
    if 'Y' in str(cdr3)[6:8]:
        
        pro = SeqRecord(translate_by_3s(Bio.Seq.Seq(str(vhseq))),
                                id='id',
                                name='id',
                                description='id')
        prot=re.sub('X','',str(pro.seq))
        
        # Check for conserved residues in HCDR2 sequence
        cdr2=prot.split(str(cdr3))[0][-45:-36]
        if 'IF' in cdr2[3:5] or 'MF' in cdr2[3:5]:
            #print(cdr2)
            return 'IGHV1-69_motif_HA_stem_associated'
        
def check_TIhead_binding(cdr3_seq,vh,vl,vlseq):
    # Check for reported IGH and IGL v gene calls
    vhs=['V4-61','V3-30','V3-7','V2-70','V4-59','V4-34']
    vls=['KV1-39']
    vh_hit=[1 for h in vhs if re.search(h,str(vh),re.IGNORECASE)]
    vl_hit=[1 for l in vls if re.search(l,str(vl),re.IGNORECASE)]
    
    # Check for conserved residues in HCDR3 sequence
    if len(vh_hit) > 0 and len(vl_hit) > 0 and len(cdr3_seq) > 6:
        if cdr3_seq[6] == 'D' or cdr3_seq[6] == 'E':
            # Translate light chain sequence
            pro = SeqRecord(translate_by_3s(Bio.Seq.Seq(str(vlseq))),
                    id='id',
                    name='id',
                    description='id')
          
            # Check for conserved residues in IGL sequence
            if re.search('NLQ',str(pro.seq)):
                return 'Trimer_interface_motif_HA_head_associated'
            
def check_Ab_signature(cdr3_seq):
    # Check for conserved residues in HCDR3 sequence
    pat_FGV_I = re.compile(r"FG[VI]")
    cdr3_seq = (cdr3_seq or "").upper()
    return 'IGHV6-1_IGHD3-3_motif_HA_stem_associated' if pat_FGV_I.search(cdr3_seq) else ''

def check_d39(d_call,cdr3_seq):
    # Check for IGHD3-9 gene call and reported sequence signatures in HCDR3 sequence
    if 'D3-9' in str(d_call):
        d3_9=re.compile('L.{1}.[YF].{1}WL')
        if d3_9.search(cdr3_seq):
            return 'IGHD3-9_motif_predominantly_HA_stem_associated'
        
def motif_in_middle(seq, motifs=("DR", "RD"), tolerance=1):
    """ Check if the reported motif is located in the middle of CDR3 """
    L = len(seq)
    center = L / 2

    for motif in motifs:
        pos = seq.find(motif)
        if pos != -1:
            motif_center = pos + (len(motif) / 2)
            if abs(motif_center - center) <= tolerance:
                return True

    return False

def check_NA(cdr3_seq,j_call,d_call,v_call,flank=8):
    
    # Check for reported IGHV and IGHD gene calls
    published_IGHD_genes=['IGHD3-22','IGHD2-2','IGHD3-10','IGHD3-3','IGHD2-15',]
    try:
        d_gene=d_call.split('*')[0]
    except:
        d_gene=''
        pass
    try:
        v_gene=v_call.split('*')[0]
    except:
        v_gene=''
        pass
        
    # Check if the conserved residues are located in the middle of HCDR3
    cdr3_seq=cdr3_seq.strip()[1:-1]
    middle=motif_in_middle(cdr3_seq)
    if middle and 'IGHJ6' in j_call and len(cdr3_seq) >= 19 and len(cdr3_seq) <= 30:
        if (
            ('IGHD5-18' in d_gene and v_gene in ['IGHV1-69', 'IGHV4-61'])
            or
            (d_gene in published_IGHD_genes)
                ):
            for m in ("DR", "RD"):
                for i in range(len(cdr3_seq)):
                    if cdr3_seq.startswith(m, i):
                        before = i
                        after = len(cdr3_seq) - (i + len(m))
                        if before >= flank and after >= flank:
                            return 'NA reactive-site binding motif'
                        
def check_qxxv(cdr3_seq,vhcall,d_call):
    # Check for the reported sequence signatures in the IGHV1-18 HCDR3 sequences
    stem_qxxv=re.compile('Q.{2}V')
    if stem_qxxv.search(cdr3_seq) and re.search('IGHV1-18',str(vhcall)):
        if re.search('D3-9',str(d_call)):
            return 'HV1-18_QXXV_motif_D-39_HA_stem_class'
        elif re.search('D2-15',str(d_call)) or re.search('D2-2',str(d_call)):
            return 'HV1-18_QXXV_motif_with_other_published_D-genes_HA_stem_class'
        
        else:
            return 'Other_HV1-18_QXXV_motif_HA_stem_class'
        

def check_anchorAbs(vhcall,vlcall,vlcdr3,vhcdr3,vhseq):
    # Check for reported IGHV and IGLV genes and reported conserved residues in the LCDR3 and HCDR2 sequences
    vh=str(vhcall).split('*')[0]
    vl=str(vlcall).split('*')[0]
    
    if vh in ['IGHV3-23','IGHV3-30','IGHV3-48',] and vl in ['IGKV3-11','IGKV3-15']:
        # Check for conserved residues in LCDR3
        if 'NWPP' in str(vlcdr3):
            
            pro = SeqRecord(translate_by_3s(Bio.Seq.Seq(str(vhseq))),
                                id='id',
                                name='id',
                                description='id')
            prot=re.sub('X','',str(pro.seq))
     
            # Check for conserved residues in HCDR2
            cdr2=prot.split(str(vhcdr3))[0][-45:-36]
            if cdr2[-1:] == 'Y':
                return 'HA_Anchor_associated'

def check_HAhead_binding_lpaf(cdr3_seq,vh,vl,vhseq):
    # Check for reported heavy and light chain V genes
    vhs=['V2-70',]
    vls=['LV3-21','LV3-9','LV1-36','LV2-24']
    vh_hit=[1 for h in vhs if re.search(h,str(vh),re.IGNORECASE)]
    vl_hit=[1 for l in vls if re.search(l,str(vl),re.IGNORECASE)]
    # Check for conserved residues in HCDR3
    if 'YGD' in cdr3_seq and len(vh_hit) > 0 and len(vl_hit) > 0:
        
        pro = SeqRecord(translate_by_3s(Bio.Seq.Seq(str(vhseq))),
                                id='id',
                                name='id',
                                description='id')
        prot=re.sub('X','',str(pro.seq))
        cdr2=prot.split(str(cdr3_seq))[0][-45:-36]
        # Check for conserved residues in HCDR2
        if 'WD' in cdr2 or 'WN' in cdr2:
            return 'H1N1_LPAF_motif_HA_head_associated'
    
def check_CH65_HAhead_binding(cdr3_seq,vh,vl,j_call):
    vd_in_middle=''
    ve_in_middle=''
    # Check for reported heavy and light chain V gene calls
    c65_vhs=['V4-59','V4-4','V4-30']
    c65_vls=['KV1-39','KV3-20','KV3-11','KV1-17','KV1-33','LV3-19','LV1-44','LV3-21','LV7-43','LV6-57']
    vh_hit=[1 for h in c65_vhs if re.search(h,str(vh),re.IGNORECASE)]
    vl_hit=[1 for l in c65_vls if re.search(l,str(vl),re.IGNORECASE)]
    # Check for reported IGHJ6 call and HCDR3 length
    if 'IGHJ6' in j_call and len(cdr3_seq) >= 19 and len(cdr3_seq) <= 23 and len(vh_hit) > 0 and len(vl_hit) > 0:
        # Check for conserved residues in HCDR3 sequence
        if 'VW' in cdr3_seq:
            seq_center = len(cdr3_seq) / 2
            tolerance = 1   
            vd_pos = cdr3_seq.find('VD')
            ve_pos = cdr3_seq.find('VE')

            vd_in_middle = vd_pos != -1 and abs((vd_pos + 1) - seq_center) <= tolerance
            ve_in_middle = ve_pos != -1 and abs((ve_pos + 1) - seq_center) <= tolerance

        if vd_in_middle or ve_in_middle:
                return 'CH65_like_motif_HA_head_associated'

def check_HA_head_lateral_binding(cdr3_seq,vh,vl,):
    # Check for reported heavy and light chain V genes and conserved residues in HCDR3    
    yxr=re.compile('Y{1}R')
    if re.search('H3-23',str(vh),re.IGNORECASE) and re.search('K1-33',str(vl),re.IGNORECASE):
        if yxr.search(str(cdr3_seq)):
            return 'H1N1_lateral_patch_motif_HA_head_associated'

```

```python
df['CH65_like_motif_HA_head_associated']=''
df['CH65_like_motif_HA_head_associated']=df.apply(lambda row: check_CH65_HAhead_binding(row['IGH_junction_aa'],row['IGH_v_call'],row['IGL_v_call'],row['IGH_j_call']),axis=1)
print("No. of CH65_like_motif_HA_head_associated sequences: ",df[df['CH65_like_motif_HA_head_associated'] == 'CH65_like_motif_HA_head_associated'].shape[0])
```

```python
## Format motif-search results by retaining rows with at least one predicted motif assignment.
motif_cols = [
    'IGHV1-18_QXXV_HA_stem_associated',
    'IGHD3-9_motif_predominantly_HA_stem_associated',
    'NA reactive-site binding motif',
    'Trimer_interface_motif_HA_head_associated',
    'CH65_like_motif_HA_head_associated',
    'IGHV6-1_IGHD3-3_motif_HA_stem_associated',
    'IGHV1-69_motif_HA_stem_associated',
    'H1N1_LPAF_motif_HA_head_associated',
    'H1N1_lateral_patch_motif_HA_head_associated',
    'HA_Anchor_associated'
]

motif_df = df[motif_cols].replace('', pd.NA)
df_result = df.loc[motif_df.notna().any(axis=1)].copy()

df_result['motif_search_results'] = (
    df_result[motif_cols]
    .fillna('')
    .agg('; '.join, axis=1)
    .str.replace(r'(;\s*)+', '; ', regex=True)
    .str.strip('; ')
)
```

```python
# Combine all predicted motif assignments into a single column.
df_result['motif_search_results']=df_result['IGHV1-69_motif_HA_stem_associated'].fillna('')+df_result['Trimer_interface_motif_HA_head_associated'].fillna('')+df_result['IGHV6-1_IGHD3-3_motif_HA_stem_associated'].fillna('')+df_result['IGHD3-9_motif_predominantly_HA_stem_associated'].fillna('')+ \
    df_result['IGHV1-18_QXXV_HA_stem_associated'].fillna('')+df_result['HA_Anchor_associated'].fillna('')+df_result['H1N1_LPAF_motif_HA_head_associated'].fillna('')+df_result['H1N1_lateral_patch_motif_HA_head_associated'].fillna('') \
    +df_result['CH65_like_motif_HA_head_associated'].fillna('')+df_result['NA reactive-site binding motif'].fillna('')
df_result['motif_search_results']=df_result['motif_search_results'].str.strip()

# Drop columns containing individual search results
df_result = df_result.drop(columns=['IGHV1-18_QXXV_HA_stem_associated',
       'IGHD3-9_motif_predominantly_HA_stem_associated',
       'NA reactive-site binding motif',
       'Trimer_interface_motif_HA_head_associated',
       'CH65_like_motif_HA_head_associated',
       'IGHV6-1_IGHD3-3_motif_HA_stem_associated',
       'IGHV1-69_motif_HA_stem_associated',
       'H1N1_LPAF_motif_HA_head_associated', 'H1N1_lateral_patch_motif_HA_head_associated','HA_Anchor_associated'])
```

```python
# Output antibody sequences containing influenza-specific motifs.
df_result_out=df_result[df_result['motif_search_results'].str.len() > 5]

### Dual motif assignments are very rare among IGHV1-69 sequences; when they occur, IGHV1-69–based predictions override CDR3-only predictions because they incorporate conserved features from both HCDR3 and HCDR2
df_result_out.loc[df_result_out['motif_search_results'].str.contains('IGHV1-69_motif_HA_stem_associatedIGHD3-9_motif_predominantly_HA_stem_associated', na=False), 'motif_search_results'] = 'IGHV1-69_motif_HA_stem_associated'
df_result_out.loc[df_result_out['motif_search_results'].str.contains('IGHV1-69_motif_HA_stem_associatedNA reactive-site binding motif', na=False), 'motif_search_results'] = 'IGHV1-69_motif_HA_stem_associated'
```
